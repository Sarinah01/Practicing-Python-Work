```python
print("Balanced Parenthesis: ")
l = []
n = input("Enter the String: ")

balanced = True

for i in n:
    if i == "(" or i == "{" or i == "[":
        l.append(i)
    elif i == ")" or i == "}" or i == "]":
        if len(l) == 0:
            balanced = False
            break
        elif i == ")" and l[-1] == "(":
            l.pop()
        elif i == "}" and l[-1] == "{":
            l.pop()
        elif i == "]" and l[-1] == "[":
            l.pop()
        else:
            balanced = False
            break

if balanced and len(l) == 0:
    print("Balanced!")
else:
    print("Unbalanced!")
```

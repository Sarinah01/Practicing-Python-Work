```python
s= input("Enter ur sentence:")
s= s.split()
d = {}

for i in range(len(s)):
    key = s[i][0]
    if key not in d:
        d[key]=[]
        d[key].append(s[i])
    else:
        d[key].append(s[i])
print(d)
   
    
```

```python
def group_by_first_letter(words):
    grouped = {}
    for word in words:
        print(word[0])
        key = word[0]
        if key not in grouped:
            grouped[key] = []
        grouped[key].append(word)
    return grouped

s = 'apple mango apple'
word_list = s.split()
result = group_by_first_letter(word_list)
print(result)
```

```python
d= {
    7 :[7,3,2],
    6:[1,5,4],
    8 :[3,2,1]
}

for i,j in d.items():
    d[i] = sorted(j)
d = dict(sorted(d.items()))
print(d)
```

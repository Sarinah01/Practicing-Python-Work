```python
from numpy import random
n = int(input("Enter how many times you want to play the game:  "))
balance=1000
for i in range(n):
    if balance >0:
        ch = int(input("Enter the num: "))
        r =  random.randint(1,7)
        if r == 3:
            if ch == r:
                print("Exacttt 3!")
                balance+=500
            else:
                print("You losttt!")
                balance-=100
        elif r < 3:
                if ch <= r:
                    print("Less than 3!")
                    balance+=100
                else:
                    print("You losttt!")
                    balance-=100
        elif r > 3:
                if ch >= r:
                    print("greater than 3!")
                    balance+=100
                else:
                    print("You losttt!")
                    balance-=100
    else:
        print("Insufficient Balance!!")
print("Your balace: ", balance)
```

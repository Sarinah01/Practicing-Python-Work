```python
print("----------OOPs------------")
class A:
    def __init__(self):
        print("Parent 1")
    def show(self):
        print("Parent show")
class B(A):
    def __init__(self):
        super().__init__()
        print("Child 1")
    def show(self):
        super().show()
        print("Child show")
b= B()
b.show()
a= A()
a.show()
 
```

```python
# multilevel , multiple
class A:
    def __init__(self):
        print("Class A")
    def show(self):
        print("Parent show")
class C:
    def __init__(self):
        print("Class  C")
    def show(self):
        print("Parent show")
class B(A, C):
    def __init__(self):
        super().__init__()
        print("Class B")
    def show(self):
        super().show()
        print("Child show")
obj = B()
b.show()
```
